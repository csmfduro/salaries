```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the data
df = pd.read_csv('salaries.csv')
```

```python
# 1 Calculate the average salary for each experience level
avg_salary_experience = df.groupby('experience_level')['salary_in_usd'].mean()
print(avg_salary_experience)
```

```python

# Plotting the average salary by Experience Level
avg_salary_experience.plot(kind='bar', color='skyblue')

# Adding title and labels
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary')

# Display the plot
plt.show()
```

```python
# 2.Are salaries significantly different across company sizes (Small, Medium, Large)?
avg_salary_company_size = df.groupby('company_size')['salary_in_usd'].mean()
print(avg_salary_company_size)
```

```python
avg_salary_company_size.plot(kind='bar', color='skyblue')

# Adding title and labels
plt.title('Average Salary across company sizes')
plt.xlabel('Company Size')
plt.ylabel('Average Salary')

# Display the plot
plt.show()
```

```python
average_salary_employment_type = df.groupby('employment_type')['salary_in_usd'].mean()
print(average_salary_employment_type)
```

```python
average_salary_employment_type.plot(kind='bar', color='skyblue')

# Adding title and labels
plt.title('Average Salary by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Average Salary')

# Display the plot
plt.show()
```

```python
#  Do fully remote jobs pay more or less than in-office jobs?
average_salary_remote = df.groupby('remote_ratio')['salary_in_usd'].mean()
print(average_salary_remote)
```

```python
average_salary_remote.plot(kind='bar', color='skyblue')

# Adding title and labels
plt.title('Average Salary Remote Ratio')
plt.xlabel('Remote Ratio')
plt.ylabel('Average Salary')

# Display the plot
plt.show()
```

```python
# What is the average salary difference between remote and non-remote employees?
average_remote = print(df.loc[df['remote_ratio'] == 100, ['remote_ratio', 'salary_in_usd']].mean())
print(df.loc[df['remote_ratio'] == 0, ['remote_ratio', 'salary_in_usd']].mean())
```

```python
#  Which countries offer the highest salaries for AI professionals?
highest_salary_country = df.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).head(15)
print(highest_salary_country)
```

```python
highest_salary_country.plot(kind='bar', color='skyblue')

# Adding title and labels
plt.title('Average Salary Country')
plt.xlabel('country')
plt.ylabel('Average Salary')

# Display the plot
plt.show()
```
